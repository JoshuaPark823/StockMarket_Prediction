```python
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle

from sklearn import linear_model
from sklearn.utils import shuffle
```

```python
data = pd.read_csv("./stock_market_data/sp500/csv/MSFT.csv")
data.tail()
```

```python
predict = "Adjusted Close"

data = data.drop("Date", axis=1)

x = np.array(data.drop([predict], axis=1))
y = np.array(data[predict])
data.tail()
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.35)

model = linear_model.LinearRegression()
model.fit(x_train, y_train)

acc = model.score(x_test, y_test)
print(acc)

with open("model/MSFT/linear_model.pickle", "wb") as f:
    pickle.dump(model, f)
```
